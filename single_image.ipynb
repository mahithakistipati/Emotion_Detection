```python
from keras.models import load_model
from keras.preprocessing.image import img_to_array
import cv2
import numpy as np
from keras.models import model_from_json

model =load_model('model.h5')
label = ['Angry','Disgust','Fear','Happy','Neutral', 'Sad', 'Surprise']

```

```python
from mtcnn import MTCNN
import cv2
from tensorflow.keras.utils import img_to_array, load_img

def predicting(image_path):
    # Load the image with the correct target size and color mode
    image = cv2.imread(image_path)
    
    # Convert image to RGB (OpenCV loads images in BGR)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Initialize MTCNN
    detector = MTCNN()

    # Detect faces in the image
    faces = detector.detect_faces(image)

    # Process each detected face
    for result in faces:
        x, y, w, h = result['box']
        face_crop = image[y:y+h, x:x+w]
        face_resized = cv2.resize(face_crop, (48, 48))

        
        if face_resized.shape[2] == 3:  # Check if the image has 3 channels
            face_resized = cv2.cvtColor(face_resized, cv2.COLOR_RGB2GRAY)

        # Convert the image to a numpy array
        feature = img_to_array(face_resized)

        # Reshape the array to the shape expected by the model (batch_size, height, width, channels)
        feature = feature.reshape(1, 48, 48, 1)
    
        # Normalize the pixel values
        return feature / 255.0

```

```python
import matplotlib.pyplot as plt
image_path = "Testing Images/lucy-nwosu.jpeg"
actual_image=cv2.imread(image_path)
actual_img_rgb = cv2.cvtColor(actual_image, cv2.COLOR_BGR2RGB)
plt.imshow(actual_img_rgb)
img = predicting(image_path)
pred = model.predict(img)
pred_label = label[pred.argmax()]
print("model prediction is ", pred_label)
```
