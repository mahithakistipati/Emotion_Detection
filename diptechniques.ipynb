```python
import os
import cv2

# Path to the FER-2013 dataset
dataset_path = 'C:/Users/mahit/OneDrive/Desktop/Emotion Detection/fer2013'


# Loading images from the dataset with support for subdirectories
def load_images_from_folder(folder):
    images = []
    for root, dirs, files in os.walk(folder):
        for filename in files:
            try:
                img = cv2.imread(os.path.join(root, filename))
                if img is not None:
                    images.append(img)
            except Exception as e:
                print(f"Error loading image '{filename}': {e}")
    return images


# Load training and testing images
train_images = load_images_from_folder(os.path.join(dataset_path, 'train'))
test_images = load_images_from_folder(os.path.join(dataset_path, 'test'))

print("Number of training images loaded:", len(train_images))
print("Number of testing images loaded:", len(test_images))
```

```python
import cv2
import os
from matplotlib import pyplot as plt
import numpy as np

source_folder = os.path.join(dataset_path, 'train/angry')
destination_folder = 'C:/Users/mahit/OneDrive/Desktop/Emotion Detection/fer2013/train_new/angry_new'

# Ensure destination directory exists
os.makedirs(destination_folder, exist_ok=True)

# Perform resizing and saving of images
for filename in os.listdir(source_folder):
    file_path = os.path.join(source_folder, filename)
    # Load the image
    original_image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
    if original_image is None:
        continue

    # Define scale factors
    scale_x = 8
    scale_y = 8

# Determine new dimensions

    new_width = int(original_image.shape[1] * scale_x)
    new_height = int(original_image.shape[0] * scale_y)

# Perform bicubic interpolation
    interpolated_image = cv2.resize(original_image, (new_width, new_height), interpolation=cv2.INTER_CUBIC)
    
    
# Define the sharpening kernel
    sharpening_kernel = np.array([[-1, -1, -1],
                                [-1,  9, -1],
                                [-1, -1, -1]])

# Apply the sharpening kernel
    sharpened_image = cv2.filter2D(interpolated_image, -1, sharpening_kernel)
    
    
    

# Perform histogram equalization
    equalized_image = cv2.equalizeHist(sharpened_image)
    
    
    
# Create a CLAHE object (Arguments are optional)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# Apply CLAHE
    enhanced_image = clahe.apply(equalized_image)
    

    # Save the interpolated image to the new directory
    save_path = os.path.join(destination_folder, filename)
    cv2.imwrite(save_path, enhanced_image)

# Just to confirm, let's check the number of images in the new directory
print("Number of images in the new 'angry_new' folder:", len(os.listdir(destination_folder)))
```

```python
import cv2
import numpy as np
import face_recognition

# Load the image
image = cv2.imread("face_image.jpg")

# Find face locations and landmarks in the image
face_locations = face_recognition.face_locations(image)
face_landmarks_list = face_recognition.face_landmarks(image, face_locations)

# Check if a face is detected
if len(face_locations) > 0:
    # Assume the first face found is the desired face
    face_landmarks = face_landmarks_list[0]
    
    # Convert landmarks to numpy array for easier manipulation
    landmarks_np = np.array(face_landmarks["chin"])

    # Calculate the rotation angle based on the angle between the eyes
    left_eye = np.mean(landmarks_np[36:42], axis=0)
    right_eye = np.mean(landmarks_np[42:48], axis=0)
    dx = right_eye[0] - left_eye[0]
    dy = right_eye[1] - left_eye[1]
    angle = np.degrees(np.arctan2(dy, dx))

    # Rotate the image
    center = ((face_locations[0][1] + face_locations[0][3]) // 2, (face_locations[0][0] + face_locations[0][2]) // 2)
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale=1)
    aligned_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]), flags=cv2.INTER_CUBIC)


# Display results
plt.figure(figsize=(10, 8))


plt.subplot(141)
plt.title('Enhanced Image (CLAHE)')
plt.imshow(enhanced_image, cmap='gray')
plt.axis('off')

plt.subplot(142)
plt.title('Aligned Image')
plt.imshow(aligned_image, cmap='gray')
plt.axis('off')

plt.show()
```
